```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets  import  make_blobs
# the make_blobs function can be generate points which follow a gaussiion distribution
```

```python
#creat data points
dataset = make_blobs(n_samples=1000,
                     n_features = 2,
                     centers = 5,
                     cluster_std=1.2,
                     random_state=0
                     )
# this datset will have points as well as clusters. we will take only the datapoints
```

```python
# printing the dataset
dataset
```

```python
data = dataset[0]
```

```python
# now we will import the sub package 'cluster'  from scipy
import scipy.cluster.hierarchy as sch
from sklearn.cluster import  AgglomerativeClustering
#Agglomerative clustering is same as hierarchical clustering
```

```python
#creating a dendogram
dendrogram=sch.dendrogram(sch.linkage(data,method = 'ward'))
```

# Wards  Procedure is a variance method  which attempts to create clusters with minimum variance.For each  cluster,the mean of all the obsevations are computed.Next for each observation the squared Euclidean distance to the cluster means is calculated .These distance are summed for all objects.

```python
#generate a scatter plot for all our observations to check how many clusters will be ideal
plt.scatter(dataset[0] [:,0],dataset[0] [:,1])
```

```python
# now we perform the actual clustering 
hcl =AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward' )
```

```python
#Predicting the cluster
y_hcl = hcl.fit_predict(data)
```

```python
#visualising our clusters:
plt.scatter(data[y_hcl ==0,0],data[y_hcl ==0,1],s=20,c='g')
plt.scatter(data[y_hcl ==1,0],data[y_hcl ==1,1],s=20,c='r')
plt.scatter(data[y_hcl ==2,0],data[y_hcl ==2,1],s=20,c='m')
plt.scatter(data[y_hcl ==0,0],data[y_hcl ==0,1],s=20,c='b')
```
